# Objective:
Calculate the monthly income.


```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Data Science/DataBases/bank_transactions.csv')
df.sample(5)
```

```python
df.describe()
```

```python
#Mean monthly income.
mean_income = df['CustAccountBalance'].mean()
print(f"The mean incoming for the customers are: {mean_income.round()}INR")
```

```python
#Median income value.
median_income = df['CustAccountBalance'].median()
print(f"The median incoming for the customers are: {median_income}INR")
```

```python
#which income value appears most frequently?
mode_income = df['CustAccountBalance'].mode()[0]
print(f"The most frequently incoming for the customers are: {mode_income}INR")
```
